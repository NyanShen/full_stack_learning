```python
from langchain_ollama import OllamaLLM

# 初始化模型
llm = OllamaLLM(
    base_url="http://192.168.7.200:11434",  # 可配置为内部服务器地址
    model="deepseek-r1:32b",
    temperature=0.8,
    num_ctx=4096  # 上下文窗口大小
)

# 构建链式调用
response = llm.invoke("如何优化Python代码性能？")
print(response)
```
